### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
#读入数据
data = pd.read_csv('CEPS.csv',encoding='gb2312')
data
```

```python
#随机抽取1000个数据
sample_data_1 = data.sample(n = 1000)
sample_data_1
```

```python
#选取两个连续变量
continuous_variables = ['sweight','cog3pl']
#计算两个连续变量的均值
variable_mean = sample_data_1[continuous_variables].mean()
print('\n ---the means of two continuous_variables in sample_data_1---\n',variable_mean)
```

```python
#重复抽取得到30个样本均值
mean_dataframe = pd.DataFrame(columns = continuous_variables)
for i in range(30):
    sample_data_2 = data.sample(n = 1000,replace = False,axis = 0)
    variable_mean = pd.DataFrame(sample_data_2[continuous_variables].mean()).T
    mean_dataframe = pd.concat([mean_dataframe,variable_mean],axis = 0)
mean_dataframe = mean_dataframe.reset_index(drop=True)
print('\n------30 means for samples------\n',mean_dataframe)
```

```python
#绘制直方图
mean_dataframe['sweight'].hist()
plt.show()
mean_dataframe['cog3pl'].hist()
plt.show()
```

```python
#计算均值和标准差
means = pd.DataFrame(mean_dataframe.mean(),columns = ['means'])
stds = pd.DataFrame(mean_dataframe.std(),columns = ['stds'])
describle = pd.concat([means,stds],axis = 1)
print('\n---the mean and std of the mean_dataframe:---\n',describle)
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码
+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数
    + 各系数的显著性检验结果
    + 对模型结果的解释
    

```python
%matplotlib inline
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
```

```python
sentinels = {'b12': [' '], 'b2301': [' '], 'c1704': [' '] ,'c12': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
```

```python
#选择变量
#b12：家里的书的数量
#b2301：父母对考试、作业的严格程度
#c1704：受班主任老师表扬的程度
#c12：目前的成绩
df1=df.sample(n=1000)
df2 = pd.DataFrame({
    'x1': df1.b12,
    'x2': df1.b2301,
    'x3': df1.c1704,
    'y':  df1.c12})
df2=df2.dropna(axis=0,how='any')
df2
```

```python
#自变量、因变量
model_x= ['x1','x2','x3']
x = df2.loc[ :,model_x].values
y=df2['y'].values
```

```python
#最小二乘法及变量的系数
model = sm.OLS(y, x)  
results = model.fit()
results.params
```

```python
print(results.summary())
```

结论：
1.模型拟合指标：自变量为学生家里书的数量；父母对考试、作业的严格程度；受班主任表扬的程度。因变量为学生目前的成绩。
2.模型的显著性检验结果：P＜0.005，因此自变量对其有显著性影响。
3.变量的系数：x1：0.1885；x2：0.5709；x3：0.3831。
4.各系数的显著性检验结果：x1、x2、x3的P值均小于0.05，因此x1、x2、x3均对因变量有显著影响。
5.对模型结果的解释：从分析结果可以看出，学生家里的书的数量、学生父母对考试和作业的严格程度、学生受班主任表扬的程度，均对学生目前的成绩有显著性影响，且均为正相关。
