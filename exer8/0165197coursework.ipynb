### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
sentinels = {'a1201': [' '], 'a1202': [' ']}
df=pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
#将CEPS.csv数据读入python
df
```

```python
a=df.sample(n=1000)#随机抽取1000个数据
a
```

```python
x1=a['a1201']
print(x1)
x2=a['a1202']
print(x2)#根据问卷从数据中挑选'a1201'和'a1202"这两个连续型变量
```

```python
x1.mean()#计算这两个连续变量的均值
```

```python
x2.mean()#计算这两个连续变量的均值
```

```python
count=0
mean_1201={}
mean_1202={}
while count<30:#重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
    df_a=df.sample(n=1000)
    mean_1201_a=df_a['a1201'].mean()
    mean_1201[count+1]= mean_1201_a
    mean_1202_a=df_a['a1202'].mean()
    mean_1202[count+1]= mean_1202_a
    count=count+1
print(mean_1201)
print(mean_1202)



```

```python
ind=np.arange(len(mean_1201.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_1201.values(), width, 
                color='B', label='a1201均值')
ax.set_ylabel('同意程度')
ax.set_title('随机抽取a1201题样本的同意程度均值')
ax.set_xticks(ind)
ax.set_xticklabels(mean_1201.keys())
ax.legend()
ax.set_ylim(1,5)

def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')



plt.show()
```

```python
%matplotlib inline
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
```

选取b18题“睡眠时间”为因变量y

选取b14题中“过去的一周里写老师布置作业的时间”，b15题中1“写家长、补习班布置作业”、2“上与课有关的补习班”、3“参加体育活动”、4“读课外书”、5“看电视”、6、“上网玩游戏”、7、“做家务或帮父母干活”分别为自变量：x1,x2,x3.x4,x5,x6,x7,x8


```python
sentinels = {'b14a1': [' '], 'b14a2': [' '],'b14b1': [' '],'b14b2': [' '],'b15a1':[' '],'b15a2':[' '],'b15b1':[' '],'b15b2':[' '],'b15c1':[' '],'b15c2':[' '],'b15d1':[' '],'b15d2':[' '],'b15e1':[' '],'b15e2':[' '],'b15f1':[' '],'b15f2':[' '],'b15g1':[' '],'b15g2':[' '],'b16a1':[' '],'b16a2':[' '],'b16b1':[' '],'b16b2':[' '],'b16c1':[' '],'b16c2':[' '],'b16d1':[' '],'b16d2':[' '],'b16e1':[' '],'b16e2':[' '],'b16f1':[' '],'b16f2':[' '],'b16g1':[' '],'b16g2':[' '],'b18a':[' '],'b18b':[' ']}
df=pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
b=df.sample(n=1000)
b
```

```python
#建立变量数据框,并将变量中的分钟转化为小时相加

data=pd.DataFrame({'x1':b.b14a1+b.b14a2/60+b.b14b1+b.b14b2/60,
                   'x2':b.b15a1+b.b15a2/60+b.b16a1+b.b16a2/60,
                  'x3':b.b15b1+b.b15b2/60+b.b16b1+b.b16b2/60,
                  'x4':b.b15c1+b.b15c2/60+b.b16c1+b.b16c2/60,
                  'x5':b.b15d1+b.b15d2/60+b.b16d1+b.b16d2/60,
                  'x6':b.b15e1+b.b15e2/60+b.b16e1+b.b16e2/60,
                  'x7':b.b15f1+b.b15f2/60+b.b16f1+b.b16f2/60,
                  'x8':b.b15g1+b.b15g2/60+b.b16g1+b.b16g2/60,
                  'y':b.b18a+b.b18b/60})
                 

data=data.dropna(axis=0,how='any')#删除缺失值
data
```

```python
#构建自变量
model_x= ['x1','x2','x3','x4','x5','x6','x7','x8']
X = data.loc[ :,model_x].values
#构建因变量
y=data['y'].values
X_model = sm.add_constant(X)#加入截距
print(X)
print(y)
print(X_model)
```

```python
model = sm.OLS(y, X)    #拟合普通最小二乘回归
results = model.fit()
results.params
```

```python
print(results.summary())#输出结果
```
