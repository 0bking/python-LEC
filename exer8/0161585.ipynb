### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False 
data=pd.read_csv('CEPS.csv',encoding='gb2312')
data
```

```python
sentinels = {'stdchn': [' '], 'stdeng': [' ']}
data= pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df=data.sample(n=1000)
x1=df['stdchn'].mean()
x2=df['stdeng'].mean()
print('1000个样本中标准化语文成绩均值为:%.2f'%x1)
print('1000个样本中标准化英语成绩均值为:%.2f'%x2)
```

```python
x1=[]
x2=[]
for i in range(1,31):
    df=data.sample(1000)
    df['stdchn']=df['stdchn'].astype('float')
    df['stdeng']=df['stdeng'].astype('float')
    x1.append(np.mean(df.stdchn))
    x2.append(np.mean(df.stdeng))
print(x1)
print('\n')
print(x2)
```

```python
fig = plt.figure()
ax1=fig.add_subplot(111)
rects=plt.bar(range(len(x1)), x1,color='black')
ax1.set_ylim(68,72)
index = range(1,31)
index2 = range(0,30)
plt.xticks(index2,index)
plt.title("30个样本组的标准化语文成绩均值直方图", fontsize=20)
plt.ylabel("成绩", fontsize=15) 
```

```python
fig = plt.figure()
ax1=fig.add_subplot(111)
rects=plt.bar(range(len(x2)), x2,color='black')
ax1.set_ylim(68,72)
index = range(1,31)
index2 = range(0,30)
plt.xticks(index2,index)
plt.title("30个样本组的标准化英语成绩均值直方图", fontsize=20)
plt.ylabel("成绩", fontsize=15) 
```

```python
i = 1
sum = 0
while i < 30:
    sum += x1[i]
    i += 1
y1 = (sum / 30)
print("30组标准化语文成绩均值的均值为：%f" %y1)
i = 1
sum = 0
while i < 30:
    sum += x2[i]
    i += 1
y2 = (sum / 30)
print("30组标准化英语成绩均值的均值为：%f" %y2)
```

```python
i = 1
sum = 0
while i < 30:
    sum += (x1[i]-y1)**2
    i += 1
z1 = (sum / 30)**0.5
print("30组标准化语文成绩均值的标准误为：%f" %z1)
i = 1
sum = 0
while i < 30:
    sum += (x2[i]-y2)**2
    i += 1
z2 = (sum / 30)**0.5
print("30组标准化英语成绩均值的标准误为：%f" %z2)
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码
+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数
    + 各系数的显著性检验结果
    + 对模型结果的解释
    

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
sentinels = {'b06': [' '],'b07': [' '],'b12': [' '],'b15d1': [' '],'b15d2': [' '],'stdchn': [' ']}
df= pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df
```

```python
data=pd.DataFrame({'x1':df.b06,
                   'x2':df.b07,
                   'x3':df.b12,
                   'x4':df.b15d1*60+df.b15d2,
                   'y':df.stdchn})
data=data.dropna(axis=0,how='any')
data
```

```python
model_x= ['x1','x2','x3','x4']
X = data.loc[ :,model_x].values
y=data['y'].values
model = sm.OLS(y,X)
results = model.fit()
print(results.summary())
```

```python
结果解释：
Y——标准化的语文成绩
X1——妈妈的受教育程度
X2——爸爸的受教育程度
X3——家里的书籍数目
X4——课外书的阅读时间
1、其中R-squared以及Adj. R-squared都为0.887，说明回归模型结果比较显著，父母的受教育程度、家里书的多少以及课外书的阅读时间能够对语文成绩做出解释。
2、Prob(Omnibus)为0.000，说明统计结果较为显著。
3、各个因变量的p值均为0.000，t值较大，说明各个因变量对Y的影响是显著的。
4、各个因变量的系数（即1.6355 3.9065 12.0945 0.0380）均为正数，说明各个因变量与自变量Y为正相关关系。
5、其中X3（家里的书多或少）的系数最大，说明家里书籍的多少对学生语文成绩的影响是最大的，在其他条件不变的情况下，每增加一本书籍，标准化语文成绩增加12.09分。
```
