```python
%matplotlib inline
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
```

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量

挑选分析所需的变量

```python
sentinels = {'b01': [' '], 'b06': [' '], 'b07': [' '], 'c08': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
```

自变量：x1-是否为独生子女
      x2-妈妈的教育水平
      x3-爸爸的教育水平
因变量：小学六年级的时候成绩排名

```python
df1=df.sample(n=1000)
shuju = pd.DataFrame({
    'x1': df1.b01,
    'x2': df1.b06,
    'x3': df1.b07,
    'y': df1.c08})
shuju=shuju.dropna(axis=0,how='any')
shuju
```

构建因变量和自变量

```python
model_x= ['x1','x2','x3']
x = shuju.loc[ :,model_x].values
y=shuju['y'].values
```

拟合最小二乘法，并给出各自变量的系数

```python
model = sm.OLS(y, x)  
results = model.fit()
results.params
```

输出检验结果

```python
print(results.summary())
```

解释分析：
1.模型拟合指标:自变量:x1-是否为独生子女;x2-妈妈的教育水平;x3-爸爸的教育水平.
               因变量:y-小学六年级的时候成绩排名.

2.模型的显著性检验结果:该模型的p值为0<0.05,所以在该模型中自变量有显著作用.

3.变量的系数:x1的系数为6.5706924,x2的系数为0.53524988,x3的系数为0.64122249.

4.各系数的显著性检验结果:在置信区间95%的情况下,x1的p值为0,x2的p值为0.036,x3的p值=0.009,所以x1、x2和x3的作用是显著的.

5.对模型结果的解释:通过该检验结果可知,x1（是否为独生子女）、x2（母亲的受教育程度）和x3（父亲的受教育程度）对班级排名y起显著作用.
