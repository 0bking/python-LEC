### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
#数据的读入
sentinels = {'tr_chn': [' '], 'tr_mat': [' ']}
df=pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df.head()
```

```python
#随机抽数
sample=df.sample(n=1000)
sample
```

```python
#计算均值
print('语文成绩的均值',sample['tr_chn'].mean())
print('数学成绩的均值',sample['tr_mat'].mean())
```

```python
count=1
mean_chn=[]
mean_mat=[]
while count<=30:
    sample=df.sample(n=1000)
    x=float(sample['tr_chn'].mean())
    y=float(sample['tr_mat'].mean())
    mean_chn.append(x)
    mean_mat.append(y)
    count=count+1
print('抽取的30次学生的平均语文成绩\n',mean_chn)
print('抽取的30次学生的平均数学成绩\n',mean_mat)
```

```python
#作直方图
fig = plt.figure()
ax=fig.add_subplot(121)
numbins=10
plt.hist(mean_chn,numbins,facecolor='blue',edgecolor='black',alpha=0.8,rwidth=0.8)
plt.xlabel('区间')
plt.ylabel('频数')
plt.title('语文成绩均值分布直方图')

ax=fig.add_subplot(122)
numbins=10
plt.hist(mean_mat,numbins,facecolor='blue',edgecolor='black',alpha=0.8,rwidth=0.8)
plt.xlabel('区间')
plt.ylabel('频数')
plt.title('数学成绩均值分布直方图')
plt.show()
```

```python
#计算均值与标准误
mean_mean_chn=np.mean(mean_chn)
std_mean_chn=np.std(mean_chn)
print('语文成绩的均值为',mean_mean_chn)
print('语文成绩的标准误为',std_mean_chn)
print('\n')
mean_mean_mat=np.mean(mean_mat)
std_mean_mat=np.std(mean_mat)
print('数学成绩的均值为',mean_mean_mat)
print('数学成绩的标准误为',std_mean_mat)
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码
+ 请报告回归方程的结果，需要包括：

```python
取a1803（1~5将其重新编码使得数值越高心情越差）作为因变量Y，探究学生心情的影响因素
自变量为X1: a17健康状况（1~5，数值越高越好）
        X2: b01是否独生（重新编码为0、1变量）
        X3: （b1701+b1702）/2：与同学外出游玩的频率(数值越高频率越高)
        X4：(b2501+b2502)/2:与父母亲的关系
        X5: b09:家庭经济条件（1~5，数值越高家庭条件越好）
        X6：c12:成绩（1~5，数值越高成绩越好）        
```

```python
sentinels = {'a1803': [' '], 'a17': [' '],'b01': [' '],'b1701': [' '],'b1702':[' '],'b2501':[' '],'b2502':[' '],'b09':[' '],'c12':[' ']}
df=pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
sample=df.sample(n=1000)
sample
```

```python
#建立变量数据框
data=pd.DataFrame({'x1':sample.a17,
                   'x2':sample.b01,
                   'x3':(sample.b1701+sample.b1702)/2,
                   'x4':(sample.b2501+sample.b2502)/2,
                   'x5':sample.b09,
                   'x6':sample.c12,
                   'y' :sample.a1803})
                 
data=data.dropna(axis=0,how='any')#删除缺失值
data
```

```python
#数据重新编码
data['x2']=data['x2'].replace([1,2],[1,0])
data['y']=data['y'].replace([1,2,3,4,5],[5,4,3,2,1])
data
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
#构建自变量
model_x= ['x1','x2','x3','x4','x5','x6']
X = data.loc[ :,model_x].values
#构建因变量
y=data['y'].values
X_model = sm.add_constant(X)#加入截距
#拟合ols回归
model = sm.OLS(y, X) 
results = model.fit()
results.params
#输出结果
print(results.summary())
```

模型拟合指标: 
   因变量y: 学生心情好坏程度（值越大心情越好）
   自变量为
        x1: 健康状况（值越大越健康）
        x2: 是否独生（0为否1为是）
        x3: 与同学外出游玩的频率
        x4：与父母的关系（值越高越亲近）
        x5: 家庭经济条件（数值越高家庭条件越好）
        x6：成绩（数值越高成绩越好） 
        
模型的显著性检验结果:
    构造F统计量，其显著性检验P值<0.01，说明在99%的把握下模型显著  

变量的系数：
    x1的系数为0.3206
    x2的系数为-0.0254
    x3的系数为0.0773
    x4的系数为0.4886
    x5的系数为0.2552
    x6的系数为0.0617
  
各系数的显著性检验结果：
    x1:p<0.05
    x2:p>0.05
    x3:p<0.05
    x4:p<0.05
    x5:p<0.05
    x6:p<0.05
   除了x2(是否为独生)的系数没有通过t检验，其余5个变量系数均通过t检验，说明学生健康状况，与朋友外出频率，与父母关系，家庭经济条件，学习成绩均对学生心情具有显著影响，并且都为正相关关系。

对模型结果的解释：
    该模型R方为0.941,说明该模型自变量对因变量的解释程度为94.1%，模型拟合效果非常好
    x1为学生健康状况，说明学生健康状况每提升一个水平（如：从不太好到一般），学生快乐程度提升0.32个水平
    x3为与朋友外出频率，说明学生与朋友外出游玩频率每提升一个水平（如：从每半年一次到每月一次），学生快乐程度提升0.073个水平
    x4为与父母的关系，说明学生与父母的关系每提升一个水平（如：从一般到很亲近），学生快乐程度提升0.4886个水平
    x5为家庭经济条件，说明学生家庭经济条件每提升一个水平（如：从中等到比较富裕），学生快乐程度提升0.2552个水平
    x6为学习成绩，说明学生学习成绩每提升一个水平（如：从中等到中上），学生快乐程度提升0.0617个水平
    总体来说，对学生心情影响最大的是学生的健康状况，与父母的关系和家庭的经济状况。越健康，家庭关系越和睦，家庭经济状况越好的孩子越开心
