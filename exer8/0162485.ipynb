### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
sentinels = {'a13': [' '], 'a14': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df
```

```python
x1=[]
x2=[]
i = 0
while i<30:
    x = df.sample(n=1000,axis=0)
    a=x['a13']
    b=x['a14']
    a = a.mean()
    b = b.mean()
    x1.append(a)
    x2.append(b)
    i += 1
fig = plt.figure()
ax1=fig.add_subplot(111)
rects=plt.bar(range(len(x1)), x1)
ax1.set_ylim(150,170)
index = range(1,31)
index2 = range(0,30)
plt.xticks(index2,index)
plt.title("随机抽样身高均值", fontsize=20)
plt.ylabel("身高（cm）", fontsize=14) 
```

```python
fig = plt.figure()
ax2=fig.add_subplot(111)
rects=plt.bar(range(len(x2)), x2)
ax2.set_ylim(90,110)
plt.xticks(index2,index)
plt.title("随机抽样体重均值", fontsize=20)
plt.ylabel("体重（斤）", fontsize=14) 
```

```python
y1 = sum(x1)/len(x1)
y2 = sum(x2)/len(x2)
print("身高均值为%f厘米" %y1)
print("体重均值为%f斤" %y2)
i = 1
sum = 0
while i < 30:
    sum += (x1[i]-y1)**2
    i += 1
z1 = (sum / 30)**0.5
print("身高标准差为%f" %z1)
i = 1
sum = 0
while i < 30:
    sum += (x2[i]-y2)**2
    i += 1
z2 = (sum / 30)**0.5
print("体重标准差为%f" %z2)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

选b12“家里的书多吗”，b2201“父母有没有检查你的作业”、b2202“父母有没有指导你的功课”分别为自变量：x1,x2,x3。
选c12“你目前的成绩”为因变量y

```python
sentinels = {'b12': [' '], 'b2201': [' '],'b2202':[' '],'c12':[' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df
```

```python
#建立变量数据框
data=pd.DataFrame({'x1':df.b12,
                  'x2':df.b2201,
                  'x3':df.b2202,
                  'y':df.c12})
data=data.dropna(axis=0,how='any')
data
```

```python
model_x= ['x1','x2','x3']
X = data.loc[ :,model_x].values
y=data['y'].values
X_model = sm.add_constant(X)
model = sm.OLS(y, X)    #拟合普通最小二乘回归
results = model.fit()
print("R^2:",results.rsquared,"\nNMSE:",1-results.rsquared)
print(results.params)
```

```python
print(results.summary())#输出所有结果
```

```python
1.模型拟合指标
    “你目前的成绩”为因变量y
    “家里的书多吗”，“父母有没有检查你的作业”，“父母有没有指导你的功课”分别为自变量x1,x2,x3。
    判定系数为0.824，说明模型解释的变差占总变差的比例较高，表明父母对孩子的辅导和家里的教育资源对目前的成绩变量有较强的解释力，模型拟合结果较好。
2.模型的显著性检验结果
    P<0.001，有显著影响
3.变量的系数
    x1:0.6557
    x2:0.2192
    x3:0.1009
    由此看出三个变量对y均呈正相关
4.各系数的显著性检验结果
    x1.p<0.05
    x2.p<0.05
    x3.p<0.05
    由此看出三个变量对y均有显著影响
5.对模型结果的解释
    可认为父母对孩子的辅导和家里的教育资源对孩子目前的成绩呈正相关
```
