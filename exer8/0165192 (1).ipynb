+ 请从中随机抽取1000个数据

+ 计算这两个连续变量的均值

+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值

```python
df = csv.reader(open('CEPS.csv','r'))
stcog = []
sweight = []
for line in df:
    stcog.append(line[9])
    sweight.append(line[6])

stcog.pop(0)
sweight.pop(0)

var1 = []
var2 = []

for i in range(0,30):
    stcog_1000 = random.sample(stcog,1000)
    sweight_1000 = random.sample(sweight,1000)
    
    sum_stcog = 0.0
    sum_sweight = 0.0

    for str in stcog_1000:
        num = float(str)
        sum_stcog += num
    stcog_mean = sum_stcog/ 1000
    var1.append(stcog_mean)
    

    for str in sweight_1000:
        num = float(str)
        sum_sweight += num
    sweight_mean = sum_sweight / 1000
    var2.append(sweight_mean)                                                  

print(var1)        
print(var2)
stcog_1000
sweight_1000 
```

+ 绘制这30个样本均值的直方图

```python
plot = plt.figure()
ax1 = plot.add_subplot(1,2,1)
plt.hist(var1, bins = 40, alpha = 0.5, color = 'b', label = ['var1'])
plt.grid(True)
plt.xlabel('mean')
plt.ylabel('frequency')
plt.title('stcog')
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
plt.hist(var2, bins = 25, alpha = 0.7, color = 'purple', label = ['var2'])
plt.grid(True)
plt.xlabel('mean')
plt.ylabel('frequency')
plt.title('sweight')
plt.show()
```

+ 计算均值的均值和标准误

```python
sum_var1 = 0
for float in var1:
    sum_var1 += float
mean_var1 = sum_var1/30
print('stcog的均值为: %s'%mean_var1)
std_var1 = np.std(var1)
print('stcog的标准差为: %s'%std_var1)

sum_var2 = 0
for float in var2:
    sum_var2 += float
mean_var2 = sum_var2/30
print('sweight的均值为： %s'%mean_var2)
std_var2 = np.std(var2)
print('sweight的标准差为: %s'%std_var2)
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码

+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数
    + 各系数的显著性检验结果
    + 对模型结果的解释

+ 步骤如下：
+ 取 c22"你希望自己读到什么程度" 做因变量y
+ 取 c1301“数学对我的未来很有帮助” 、c1302“语文对我的未来很有帮助”、b09“ 家庭经济状况”、c12"目前成绩排名"、b30“父母对你的学业成绩要求”做自变量

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd

sentinels = {'c22': [' '], 'c1302': [' '], 'c1301': [' '], 'b09': [' '],'b30': [' '],'c12': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
sdf=df.sample(n=1000)
sdf
```

```python
data1= pd.DataFrame({
    'y': sdf.c22,
    'x1': sdf.c1302,
    'x2': sdf.c1301,
    'x3': sdf.b30,
    'x4': sdf.c12,
    'x5': sdf.b09
})
#构建自变量
data1=data1.dropna(axis=0,how='any')#删除缺失值
model_x = ['x1','x2','x3','x4','x5']
X = data1.loc[ :,model_x].values
#构建因变量
y = data1['y'].values
#加入截距项
X_model = sm.add_constant(X)
print(X)
print(y)
print(X_model)
model = sm.OLS(y, X) 
#拟合ols回归
results = model.fit()
results.params

#输出结果
print(results.summary())
```

 + 模型拟合指标：
   + 因变量 y ：数学对我的未来有帮助
   + 自变量：
      + x1:语文对我的未来很有帮助
      + x2:数学对我的未来很有帮助
      + x3:父母对你的学业成绩要求
      + x4:目前成绩排名
      + x5:家庭经济状况
 + 模型的显著性检验结果
   + P=0.000，有显著影响
 + 变量的系数（表中coef对应值）

    + x3:0.0202
    + x4:0.6924
    + x5:0.7634
 + 各系数的显著性检验结果（表中P对应值）
   + x1:p<0.05
   + x2:p<0.05
   + x3:p>0.05
   + x4:p<0.05
   + x5:p<0.05
   + 由此看出四个变量对y均有显著影响，x3：父母对成绩要求影响不显著
五个自变量中与因变量之间都成正相关，其中，目前成绩排名和家庭经济状况对学生的学业期望有很重要的影响，成绩好的同学会希望自己能读到更高的学历，家庭条件较好的同学也会选择继续深造，而本身成绩不好的学生，或者家境困难的学生，更容易选择放弃学业。
