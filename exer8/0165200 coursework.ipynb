+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值

```python
### 选取 c08 c19(六年级成绩班级排名与最好朋友数量)两个变量
data = {'c08':df2.c08, 
        'c19': df2.c19}
frame = pd.DataFrame(data)
frame
frame.mean()
```

+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图

```python
c08_mean = []
c19_mean = []

i = range(30)
for i in range(30):
    sample = df.sample(1000)    #随机抽取样本
    x1 = sample['c08'].mean()
    x2 = sample['c19'].mean()
    c08_mean.append(x1)
    c19_mean.append(x2)

print("c08的30个样本均值",c08_mean)
print("c19的30个样本均值",c19_mean)
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))

grade = c08_mean
friends = c19_mean

ax0.hist(grade, 10, histtype='stepfilled', facecolor='skyblue')
ax0.set_title('六年级成绩班级排名')
ax0.set_xlabel('排名')
ax0.set_ylabel('频数')

ax1.hist(friends, 15, histtype='stepfilled', facecolor='pink')
ax1.set_title('你有几个最好的朋友')
ax1.set_xlabel('数量')
ax1.set_ylabel('频数')

fig.tight_layout()
plt.show()
```

+ 计算均值的均值和标准误

```python
c08_mean = array(c08_mean)
c19_mean = array(c19_mean)

print('c08的样本均值的均值为：', c08_mean.mean())
print('c08的样本均值的标准误为：', c08_mean.std())
print('\n')
print('c19的样本均值的均值为：', c19_mean.mean())
print('c19的样本均值的标准误为：', c19_mean.std())
```
