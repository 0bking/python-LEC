### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False 
```

```python
sentinels = {'a1201': [' '], 'a1804': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df
```

```python
a=df.sample(n=1000,axis=0)#随机抽取1000个数据
```

```python
x1=a['a1201']
print(x1)
x2=a['a1804']
print(x2)
#根据问卷从数据中挑选'a1201'和'a1804"这两个连续型变量
```

```python
x1.mean()#计算连续型变量的均值
```

```python
x2.mean()#计算连续型变量的均值
```

```python
count=0
mean_1201={}
mean_1804={}
while count<30:#重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
    df_a=df.sample(n=1000)
    mean_1201_a=df_a['a1201'].mean()
    mean_1201[count+1]= mean_1201_a
    mean_1804_a=df_a['a1804'].mean()
    mean_1804[count+1]= mean_1804_a
    count=count+1
print(mean_1201)
print(mean_1804)
```

```python
ind=np.arange(len(mean_1201.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_1201.values(), width, 
                color='R', label='a1201样本均值')
ax.set_ylabel('样本均值')
ax.set_title('随机抽取a1201题的样本均值')
ax.set_xticks(ind)
ax.set_xticklabels(mean_1201.keys())
ax.legend()
ax.set_ylim(1,4)

def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')



plt.show()
```

```python
ind=np.arange(len(mean_1804.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_1804.values(), width, 
                color='B', label='a1804样本均值')
ax.set_ylabel('样本均值')
ax.set_title('随机抽取a1804题的样本均值')
ax.set_xticks(ind)
ax.set_xticklabels(mean_1804.keys())
ax.legend()
ax.set_ylim(1,2)

def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')



plt.show()
```

```python
a1201=[]
for i in mean_1201.values():
    a1201.append(i)
    
a1804=[]
for i in mean_1804.values():
    a1804.append(i)
    
average={'1201':a1201,
         '1804':a1804,
        }
frame = pd.DataFrame(average,index=mean_1201.keys())
#计算30个样本均值的均值和标准误
print('12题第1问与18题第4问的样本均值的均值如下：\n')
print(frame.mean())
print('\n12题第1问与18题第4问的标准误如下：\n')
print(frame.std())
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码
+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数
    + 各系数的显著性检验结果
    + 对模型结果的解释
    

```python
%matplotlib inline
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
挑选c12题“你目前的成绩”为因变量y
挑选b6题中“你妈妈的教育水平”，b7题中“你爸爸的教育水平”、b9题中“目前你家经济条件”分别为自变量：x1,x2,x3.
```

```python
sentinels = {'c12': [' '], 'b06': [' '],'b07':[' '],'b09':[' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df
```

```python
#建立变量数据框
data=pd.DataFrame({'x1':df.b06,
                  'x2':df.b07,
                  'x3':df.b09,
                  'y':df.c12})
data=data.dropna(axis=0,how='any')
data
```

```python
#构建自变量
model_x= ['x1','x2','x3']
X = data.loc[ :,model_x].values
#构建因变量
y=data['y'].values
#加入截距项
X_model = sm.add_constant(X)
print(X)
print(y)
print(X_model)
```

```python
import statsmodels.api as sm
model = sm.OLS(y, X)    #拟合普通最小二乘回归
results = model.fit()
print("R^2:",results.rsquared,"\nNMSE:",1-results.rsquared)
print(results.params)
```

```python
print(results.summary())#输出所有结果
```

```python
1.模型拟合指标
“你目前的成绩”为因变量y
“你妈妈的教育水平”，“你爸爸的教育水平”，“目前你家经济条件”分别为自变量x1,x2,x3。
判定系数为0.869，说明模型解释的变差占总变差的比例较高，表明父母的教育水平和家庭经济条件对学生目前的成绩变量有较强的解释力，模型拟合结果较好。

2.模型的显著性检验结果
    F统计量和Omnibus统计量的P值都近似为0，说明自变量的作用显著。

3.变量的系数
x1的系数为0.0465，x2的系数为0.0877，x3的系数为0.8188，说明三个自变量与因变量都成正相关。

4.各系数的显著性检验结果
x1,x2,x3的p值均小于0.05，拒绝自变量和学生目前的成绩不相关的假设，即这三个自变量都是统计显著的。说明这3个自变量在模型里都是有意义的。

5.对模型结果的解释
由于所选自变量教育水平在问卷数据中数字越大代表的教育水平越高，经济条件在问卷数据中数字越大代表的经济条件越好，所以此模型未经过哑变量编码处理。
Jarque-Bera的P值接近于0，表明数据服从正态分布。Durbin-Watson检验的值为1.669，表明残差服从正态分布，各变量之间是相互独立的，不存在自相关。
线性回归模型为y=0.0465x1+0.0877x2+0.8188x3
由coef栏可知，x1每变动一个单位，因变量变动0.0465个单位；x2每变动一个单位，因变量变动0.0877个单位；x3每变动一个单位，因变量变动0.8188个单位。
这三个自变量中，家庭经济条件对学生目前的成绩影响最大。
```
