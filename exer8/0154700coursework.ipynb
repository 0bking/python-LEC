### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
#读取数据
data = pd.read_csv(r'CEPS.csv',engine='python')
data
```

```python
#重复30次随机抽取1000个数据并计算均值
mean = []
for _ in range(30):
    index = np.random.randint(1,len(data),1000)
    data_1000 = data.iloc[index,:]
    var = data_1000.loc[:,['sweight','stcog']]    
    temp = np.array(var.values.tolist())           # tolist()  将数组或者矩阵转换成列表 
    mean.append(temp.mean())
mean
```

```python
#绘制均值的直方图
df_mean = pd.DataFrame(mean)
df_mean.hist(grid=False)
```

```python
#计算均值的均值
np.mean(mean)
```

```python
#计算均值的标准差
np.std(mean)
```

```python
# 读取数据
# b14a1 周一到周五每天花费多少小时学习
# b14b1 周末每天花费多少小时学习
conti_var = {'b14a1':[' '],'b14b1':[' ']}
data = pd.read_csv('CEPS.csv',encoding='gb2312',na_values=conti_var)
data1 = data.sample(n=1000)
```

```python
# 计算均值
mean_mon_fri = data1['b14a1'].mean()
mean_mon_fri
```

```python
# 计算均值
mean_sat_sun = data1['b14b1'].mean()
mean_sat_sun
```

```python
# 重复随机抽取30次并计算样本均值
count = 0
mean_mon_fri = {}
mean_sat_sun = {}
while count<30:
    data2 = data.sample(n=1000)
    mean_mon_fri2 = data2['b14a1'].mean()
    mean_mon_fri[count+1] = mean_mon_fri2 
    
    mean_sat_sun2 = data2['b14b1'].mean()
    mean_sat_sun[count+1] = mean_sat_sun2 
    count=count+1
print(mean_mon_fri)
print(mean_sat_sun)
```

```python
# 绘制b14a1直方图
i = np.arange(len(mean_mon_fri.values()))
width=0.5
fig,ax = plt.subplots()
rects1 = ax.bar(i-width/2,mean_mon_fri.values(),width,color='r',label='mean_mon_fri')
ax.set_title('mean_mon_fri',fontdict=None,loc='center')
ax.set_ylabel('learning_hour')
ax.set_xticks(i)
ax.set_xticklabels(mean_mon_fri.keys())
ax.legend()
ax.set_ylim(0,3)

def autolabel(rects,xpos='center'):
    xpos = xpos.lower()
    ha = {'center':'center','right':'left','left':'right'}
    offset = {'center':1,'right':1,'left':1}
    
    for rect in rects:
        height=rect.get_height()
        ax.text(rect.get_x()+rect.get_width()*offset[xpos],2*height,'{}'.format(height),ha=ha[xpos],va='bottom')
plt.show()
```

```python
# 绘制b14b1直方图
i = np.arange(len(mean_sat_sun.values()))
width=0.5
fig,ax = plt.subplots()
rects1 = ax.bar(i-width/2,mean_sat_sun.values(),width,color='r',label='mean_sat_sun')
ax.set_title('mean_sat_sun',fontdict=None,loc='center')
ax.set_ylabel('learning_hour')
ax.set_xticks(i)
ax.set_xticklabels(mean_sat_sun.keys())
ax.legend()
ax.set_ylim(0,4)

def autolabel(rects,xpos='center'):
    xpos = xpos.lower()
    ha = {'center':'center','right':'left','left':'right'}
    offset = {'center':1,'right':1,'left':1}
    
    for rect in rects:
        height=rect.get_height()
        ax.text(rect.get_x()+rect.get_width()*offset[xpos],2*height,'{}'.format(height),ha=ha[xpos],va='bottom')
plt.show()
```

```python
# 计算样本均值的均值和标准误
amon_fri = []
for i in mean_mon_fri.values():
    amon_fri.append(i)
    
asat_sun = []
for i in mean_sat_sun.values():
    asat_sun.append(i) 
    
expectation ={'mon_fri':amon_fri,
               'sat_sun':asat_sun,
              }

frame = pd.DataFrame(expectation,index=mean_mon_fri.keys())
print('样本均值的均值如下：\n')
print(frame.mean())
print('\n样本均值的标准误如下：\n')
print(frame.std())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#读取数据
func = {'c12':[' '],'c25':[' '],'b33':[' '],'b09':[' '],'b14a1':[' '],'b14b1':[' '],'b2301':[' ']}
data = pd.read_csv('CEPS.csv',encoding='gb2312',na_values=func)
data
```

```python
#建立数据框
data1 = pd.DataFrame({'x1':data.c25,
                     'x2':data.b33,
                     'x3':data.b09,
                     'x4':data.b14a1,
                     'x5':data.b14b1,
                     'x6':data.b2301,
                     'y':data.c12})
data1 = data1.dropna(axis=0,how='any')
data1
```

```python
#构建自变量
model_x = ['x1','x2','x3','x4','x5','x6']
X = data1.loc[ :,model_x].values
#构建因变量
y = data1['y'].values
#加入截距项
X_model = sm.add_constant(X)
print(X)
print(y)
print(X_model)
```

```python
# 拟合普通最小二乘回归
model = sm.OLS(y,X)
results = model.fit()
results.params
```

```python
#输出结果
print(results.summary())
```

```python
###  1.模型拟合指标
    # c12---目前学习成绩（因变量y）
    # c25---自己对自己未来的信心（自变量x1）
    # b33---对父母对自己的教育期望所感到的压力（自变量x2）
    # b09----目前家庭经济条件（自变量x3）
    # b14（b14a1，b14b1）----每天花费多少小时学习（自变量x4，x5）
    # b2301----父母对自己作业和考试上管的严（自变量x6）
###  2.模型显著性检验结果
    # 由p值可以看出所有自变量对因变量有显著影响
###  3.各变量的系数
    #  x1    0.4994（影响最大）
    #  x2   -0.0222(对成绩产生负影响)
    #  x3    0.3275
    #  x4    0.0124
    #  x5    0.0173
    #  x6    0.2145
### 4.各系数的显著性检验结果
    #  x1-x6的p值均小于0.05，因此拒绝原假设，即所有的自变量均对因变量有显著影响
```
