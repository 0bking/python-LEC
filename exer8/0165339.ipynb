```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
sentinels = {'a1203': [' '], 'a1803': [' ']}#a1203: 就算功课需要花好长时间才能做完，我仍然会不断地尽力去做 1803: 在过去的七天内，你是否有以下感觉：不快乐
df=pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)#读入数据+缺失值处理
df
```

+ 请从中随机抽取1000个数据

```python
a=df.sample(n=1000)
a
```

+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）

```python
x1=a['a1203']
print(x1)
x2=a['a1803']
print(x2)
```

+ 计算这两个连续变量的均值

```python
x1.mean()#“就算功课需要花好长时间才能做完，我仍然会不断地尽力去做”均值
```

```python
x2.mean()#“在过去的七天内，你是否有以下感觉：不快乐”均值
```

+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值

```python
count=0
mean_a1203={}
mean_a1803={}
while count<30:#重复30次
    df_a=df.sample(n=1000)#随机抽取
    mean_a1203_a=df_a['a1203'].mean()#计算均值
    mean_a1203[count+1]= mean_a1203_a
    mean_a1803_a=df_a['a1803'].mean()
    mean_a1803[count+1]= mean_a1803_a
    count=count+1
print(mean_a1203)
print(mean_a1803)
```

+ 绘制这30个样本均值的直方图

```python
ind=np.arange(len(mean_a1203.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_a1203.values(), width, 
                color='black', label='是否愿意尽力完成作业样本均值')
ax.set_ylabel('是否愿意尽力完成作业均值')
ax.set_title('随机是否愿意尽力完成作业样本均值')
ax.set_xticks(ind)
ax.set_xticklabels(mean_a1203.keys())
ax.legend()
ax.set_ylim(1,4)
def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')



plt.show()
```

```python
ind=np.arange(len(mean_a1803.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_a1803.values(), width, 
                color='blue', label='过去七天感到不快乐样本均值')
ax.set_ylabel('过去七天感到不快乐均值')
ax.set_title('随机过去七天感到不快乐样本均值')
ax.set_xticks(ind)
ax.set_xticklabels(mean_a1803.keys())
ax.legend()
ax.set_ylim(1,5)
def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')



plt.show()
```

+ 计算均值的均值和标准误

```python
a1203=[]#转换为列表形式
for i in mean_a1203.values():
    a1203.append(i)
    
a1803=[]
for i in mean_a1803.values():
    a1803.append(i)
    
average={'愿意程度':a1203,
         '不快乐':a1803,
        }
frame = pd.DataFrame(average,index=mean_1203.keys())
#计算30个样本均值的均值和标准误
print('完成作业愿意程度和不快乐程度样本均值的均值如下：\n')
print(frame.mean())
print('\n')
print('完成作业愿意程度和不快乐程度样本均值的标准误如下：\n')
print(frame.std())
```

建立回归方程步骤：

```python
%matplotlib inline
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
```

```python
sentinels = {'c12': [' '], 'b01': [' '], 'b06': [' '], 'b07': [' '], 'b1001': [' '], 'b1002': [' '], 'b1003': [' '], 'b29': [' '], 'b35': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df2=df.sample(n=1000)   #随机抽取2000个样本数据
df
```

```python
data = pd.DataFrame({
    'x1': df2.b01,
    'x2': df2.b06,
    'x3': df2.b07,
    'x4': df2.b1001,
    'x5': df2.b1002,
    'x6': df2.b1003,
    'x7': df2.b29,
    'x8': df2.b35,
    'y': df2.c12})#自变量与因变量构建数据框

data=data.dropna(axis=0,how='any') 
data
```

```python
model_x= ['x1','x2','x3','x4','x5','x6','x7','x8']
X = data.loc[ :,model_x].values#构建自变量

y=data['y'].values#构建因变量

X_model = sm.add_constant(X)#加上截距
print(X)
print(y)
print(X_model)
```

```python
model = sm.OLS(y, X)    #拟合普通最小二乘回归
results = model.fit()
results.params
print(results.summary())
```
