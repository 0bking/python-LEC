```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
#导入数据
df = pd.read_csv(r'C:\Users\asus\Desktop\新建文件夹 (3)\python-LEC\CEPS.csv',engine='python', na_values=sentinels)
df1=df.sample(n=1000)   #随机抽取1000个数据
df1
```

```python
sentinels = {'b32': [' '], 'c1706': [' '], 'c12': [' '], 'c25': [' ']}
data = pd.DataFrame({
    'x1': df1.b32,
    'x2': df1.c1706,
    'x3': df1.c12,
    'y': df1.c25})#选择变量
data=data.dropna(axis=0,how='any')  #删除缺失值所在样本

model_x= ['x1', 'x2','x3']
X = data.loc[ :,model_x].values
y=data['y'].values#构造方程
```

```python
model = sm.OLS(y, X)    #拟合普通最小二乘回归
results = model.fit()
```

```python
results.params  #各变量系数
```
