```python

data4.x0 = pd.to_numeric(data4.x0, errors='coerce' )
data4.x1 = pd.to_numeric(data4.x1, errors='coerce' )
data4.y = pd.to_numeric(data4.y, errors='coerce' )
import patsy
y, X = patsy.dmatrices('y ~ x0 + x1', data4)
X
```

```python
coef, resid, _, _ = np.linalg.lstsq(X, y)
```

```python
coef
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model = sm.OLS(y, X)
```
