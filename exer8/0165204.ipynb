### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
```

```python
#将CEPS.csv读入python
#挑选学生"目前的身高"和"目前的体重"两个随机变量
sentinels = {'a13': [' '], 'a14': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df1 = df.sample(n=1000,axis=0) #随机抽取1000个数据
df1
```

```python
#计算两个连续变量的均值
mean_a13 = df1['a13'].mean()
print('学生目前身高的均值为%.2f厘米'%mean_a13)
mean_a14 = df1['a14'].mean()
print('学生目前体重的均值为%.2f斤'%mean_a14)
```

```python
#重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
count=0
mean_height={}
mean_weight={}

while count<30:
    df2=df.sample(n=1000,axis=0)
    mean_a13height=df2['a13'].mean()
    mean_height[count+1]=mean_a13height    
    mean_a14weight=df2['a14'].mean()    
    mean_weight[count+1]=mean_a14weight
    count=count+1

print('30个身高均值为：')
print(mean_height)
print('\n30个体重均值为：')
print(mean_weight)
```

```python
#绘制这30个身高均值的直方图
ind = np.arange(len(mean_height.values()))
width = 0.4

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_height.values(), width, color='c')

ax.set_title('30个身高均值的直方图',fontsize=20)
ax.set_xticks(ind)
ax.set_xticklabels(mean_height.keys())
ax.legend()
ax.set_ylabel('单位：厘米',fontsize=13)
ax.set_ylim(150,170)

plt.show()
```

```python
#绘制这30个体重均值的直方图
ind = np.arange(len(mean_weight.values()))
width = 0.4

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_weight.values(), width, color='m')

ax.set_title('30个体重均值的直方图',fontsize=20)
ax.set_xticks(ind)
ax.set_xticklabels(mean_weight.keys())
ax.legend()
ax.set_ylabel('单位：斤',fontsize=13)
ax.set_ylim(90,110)

plt.show()
```

```python
#计算均值的均值和标准误
height=[]
for i in mean_height.values():
    height.append(i)
    
weight=[]
for i in mean_weight.values():
    weight.append(i)
    
average={'身高':height,
         '体重':weight}
df3 = pd.DataFrame(average,index=mean_height.keys())

print('身高均值和体重均值的均值为：')
print(df3.mean())
print('\n身高均值和体重均值的标准误为：')
print(df3.std())
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码
+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数
    + 各系数的显著性检验结果
    + 对模型结果的解释
    

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
sentinels = {'c12': [' '], 'b12': [' '], 'b13': [' '], 'c25': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df4 = df.sample(n=1000,axis=0)
df4
```

```python
#建立回归方程
equation = pd.DataFrame({'y': df4.c12,
                         'x1': df4.b12,
                         'x2': df4.b13,
                         'x3': df4.c25})
equation = equation.dropna(axis=0,how='any')

Y = equation['y'].values
model_x = ['x1', 'x2','x3']
X = equation.loc[ :,model_x].values

model = sm.OLS(Y, X)
```

```python
results = model.fit()
```

```python
#输出自变量的系数
results.params
```

```python
#输出回归模型的结果
print(results.summary())
```

+ 模型拟合指标
    + 因变量y为“C12.你目前的成绩在班里处于：1.不好 2.中下 3.中等 4.中上 5.很好”
    + 自变量x1为“B12.你家里的书多吗？（不包括课本和杂志）1.很少 2.比较少 3.一般 4.比较多 5.很多”
    + 自变量x2为“B13.你家里有电脑和网络吗？0.都没有 1.有电脑、无网络 2.有电脑和网络”
    + 自变量x3为“C25. 你对自己的未来有没有信心？1.根本没有信心 2.不太有信心 3.比较有信心 4.很有信心”
+ 模型的显著性检验结果
    + p值近似等于0，说明模型中自变量对因变量的影响十分显著 
+ 变量的系数
    + x1 0.2157
    + x2 0.0273
    + x3 0.7055
+ 各系数的显著性检验结果
    + x1 p值接近0，影响显著
    + x2 p值=0.514，拒绝原假设，影响不显著
    + x3 p值接近0，影响显著
+ 对模型结果的解释
    + 通过构造出的线性回归模型可以看出：
    + 学生家里是否有电脑和网络对成绩的影响甚小
    + 学生家里课外书的数量对成绩有一定的影响，且影响为正向的
    + 学生的自我评估（信心、期待值）对成绩的影响较大，说明学生对自己的信心程度越强，学习成绩会更高。因此树立自信对个人发展是十分必要且重要的
