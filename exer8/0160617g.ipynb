### 回归分析


+ 研究身高的影响因素
    因变量Y：身高
    自变量：x1:睡眠时间  x2:家庭经济条件  x3:体重

```python
%matplotlib inline
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
sentinels = {'a13': [' '], 'b18a': [' '], 'b09': [' '], 'a14': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
dfe=df.sample(n=1000)  
```

```python
data=pd.DataFrame({'x1':dfe.b18a,
                  'x2':dfe.b09,
                  'x3':dfe.a14,
                  'y':dfe.a13})
data=data.dropna(axis=0,how='any')
data
```

```python
model_x= ['x1', 'x2','x3']
X = data.loc[ :,model_x].values
y=data['y'].values
model = sm.OLS(y, X)  
results = model.fit()
```

```python
results.params 
```

```python
print(results.summary())
```

+ 各系数的显著性检验结果:
   + x1:p<0.05
   + x2:p<0.05
   + x3:p<0.05
以此可知，三个变量对被解释变量都有显著影响。
模型结果的实际解释:
    从各个变量的系数估计可以得出，睡眠时间每增加一个小时身高增加约6.6cm，家庭经济条件每提升一个档次身高增加16.5cm，体重每增加0.59斤身高提高0.6cm。
