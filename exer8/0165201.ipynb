### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
#根据问卷从数据中挑选A13（身高）、A14（体重）两个连续型变量并计算这两个连续变量的均值
height_mean=df1['a13'].mean()
weight_mean=df1['a14'].mean()
print('身高均值为：%.2f厘米'%height_mean)
print('体重均值为：%.2f斤'%weight_mean)
```

```python
df  = pd.read_csv("CEPS.csv",encoding="gb2312")
df.dropna(axis=0, how='any')    # 删除有缺失值的行
sample = DataFrame()
```

```python
#重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
height_mean_list = []
weight_mean_list = []
rand = [randint(1,18000) for i in range(30)]
for i in rand:
    sample = df[i:i+1000]
    height = sample['a13']
    height = to_numeric(height, errors='coerce').fillna(0)
    height_mean = height.mean()
    height_mean_list.append(height_mean)

    weight = sample['a14']
    weight = to_numeric(weight, errors='coerce').fillna(0)
    weight_mean = weight.mean()
    weight_mean_list.append(weight_mean)

print("height的30个样本均值",height_mean_list)
print("weight的30个样本均值",weight_mean_list)
```

```python
#绘制这30个样本均值的直方图

plt.subplot(121)
plt.hist(height_mean_list, bins=10,normed=1, facecolor="yellow", edgecolor="black", alpha=0.7)
# 显示横轴标签
plt.xlabel("range of avg")
# 显示纵轴标签
plt.ylabel("frequency")
# 显示图标题
plt.title("height")
plt.subplot(122)
plt.hist(weight_mean_list, bins=10,normed=1, facecolor="blue", edgecolor="black", alpha=0.7)
# 显示横轴标签
plt.xlabel("range of avg")
# 显示纵轴标签
plt.ylabel("frequency")
# 显示图标题
plt.title("weight")
plt.show()
```

```python
#计算均值的均值和标准误
height_mean_list = array(height_mean_list)
weight_mean_list = array(weight_mean_list)
height_a = height_mean_list.mean()
print("身高的样本均值为：",height_a)
height_std = height_mean_list.std(ddof=1)  
print("身高的标准误为：",height_std)
weight_a = weight_mean_list.mean()   
print("体重的样本均值为：",weight_a)
weight_std = weight_mean_list.std(ddof=1)
print("体重的标准误为：",weight_std)
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义

+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
```

```python
#将object类型的str格式转换为float型
df = df.convert_objects(convert_numeric=True)
df
```

```python
#把经济水平b09作为因变量
#把妈妈的教育水平b06作为自变量x1，爸爸的教育水平b07作为自变量x2，母亲的工作b08a作为自变量x3，母亲的工作b08b作为自变量x4
data = pd.DataFrame({
    'x1': df.b06,
    'x2': df.b07,
    'x3': df.b08a,
    'x4': df.b08b,
    'y' : df.b09})

###处理缺失值
data=data.dropna(axis=0,how='any') 
data
```

```python
#构建自变量
model_x= ['x1','x2','x3','x4']
X = data.loc[ :,model_x].values
```

```python
#构建因变量
y=data['y'].values
```

```python
#加入截距
X_model = sm.add_constant(X)
```

```python
#拟合普通最小二乘回归
model = sm.OLS(y, X)
results = model.fit()
results.params
```

```python
#输出回归的结果
print(results.summary())
```

    + 模型的显著性检验结果
