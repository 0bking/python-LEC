### 饼图
对问题“你是独生子女吗”(b01)的回答有“是”和“否”两种回答，相应的数字分别是1和2。请画一个饼图反映二者的比例。

```python
x1=df['b01']
x=[]
for x2 in x1:
    x.append(float(x2))
x3=x.count(1)
x4=x.count(2)
sizes = [x3,x4]
colors=['black','grey']
labels =[ '是', '否']
sizes = [x3,x4]
explode = (0, 0.1)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels,colors=colors, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal') 
plt.title('是否独生子女比例图')
plt.show()
```
