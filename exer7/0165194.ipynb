### 散点图
反映期中考试标准化成绩语文（stdchn）和期中考试标准化成绩数学（stdmat）的相关关系

```python
x1=df['stdchn']
x=[]
for x2 in x1:
    x.append(float(x2))
y1=df['stdmat']
y=[]
for y2 in y1:
    y.append(float(y2))

fig=plt.figure()
ax1=fig.add_subplot(111)
plt.title('标准化语文、数学成绩相关散点图', fontsize=15)
plt.xlabel('语文标准化成绩', fontsize=15)
plt.ylabel('数学标准化成绩', fontsize=15)
plt.scatter(x, y,color='b',label='对应分数')


xminorLocator = MultipleLocator(10)
yminorLocator = MultipleLocator(10)
ax1.xaxis.set_major_locator(xminorLocator)
ax1.yaxis.set_major_locator(yminorLocator)

plt.show()
```

### 饼图
对问题“你是独生子女吗”(b01)的回答有“是”和“否”两种回答，相应的数字分别是1和2。请画一个饼图反映二者的比例。

```python
x1=df['b01']
x=[]
for x2 in x1:
    x.append(float(x2))
x3=x.count(1)
x4=x.count(2)
sizes = [x3,x4]


labels = '是', '否'
sizes = [x3,x4]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('你有独生子女吗')

plt.show()
```

### 直方图
反映变量“每天晚上睡多长时间-小时”(b18a)的分布情况。

```python
x1=df['b18a']
x=[]
for x2 in x1:
    x2=float(x2)
    x.append(x2)
    
x3=0
x4=0
x5=0
for i in x:
    if i>10:
        x5+=1
    elif i>6:
        x4+=1
    else:
        x3+=1

time = (x3,x4,x5)


ind = np.arange(len(time))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, time, width,
                color='SkyBlue')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('所占人数')
ax.set_title('每天晚上睡多长时间-小时')
ax.set_xticks(ind)
ax.set_xticklabels(('6—', '7-10', '11+'))
ax.legend()


def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.57, 'left': 0.43}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')



plt.show()
```

### 柱图
