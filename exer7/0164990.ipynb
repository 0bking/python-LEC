```python
df = pd.read_csv('CEPS.csv',encoding='gb2312')
```

```python
df["stdchn"]=df["stdchn"].apply(lambda x: np.NaN if str(x).isspace() else x)
df["stdmat"]=df["stdmat"].apply(lambda x: np.NaN if str(x).isspace() else x)
df["stdchn"]=df["stdchn"].astype('float')
df["stdmat"]=df["stdmat"].astype('float')

fig = plt.figure()  
fig1 = fig.add_subplot(111)  
fig1.scatter(df['stdchn'],df['stdmat'])
fig1.set_xlabel('stdchn',fontsize=15)
fig1.set_ylabel('stdmat',fontsize=15)

fig1.set_title(u'散点图',fontproperties=font)
plt.tick_params(axis='both',which='major',labelsize=15)
plt.axis([0,150,0,150])
```
